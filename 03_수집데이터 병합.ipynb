* 데이터 프레임의 '접수완료 일시'의 type을 datetime 으로 변경

```python
df['접수완료일시']=pd.to_datetime(df['접수완료일시'],format='%Y-%m-%d %H:%M:%S')
df.info()
```

```python
df['date']=df['접수완료일시'].dt.date
df.head()
```

```python
df=pd.get_dummies(df, columns=['신고자성별'])
df=df.rename(columns={'신고자성별_1.0':'신고_남','신고자성별_2.0':'신고_여','신고자성별_3.0':'신고_불상'})
df.head()
```

```python
df['cnt']=1
temp=df.groupby(by='date')['cnt','신고_남','신고_여','신고_불상'].sum().reset_index()
temp=temp.rename(columns={'cnt':'신고건수'})
temp.head()
```

```python
temp.info()
```

```python
temp['date']=pd.to_datetime(temp['date'],format='%Y-%m-%d')
temp.info()
```

```python
template=pd.DataFrame(pd.date_range('20200101','20230118',freq='D'), columns=['date'])
template.head(3)
```

```python
template.tail(3)
```

```python
template=template.merge(temp,on='date',how='left')
template.head()
```

* 20개 날짜에서 신고건수가 0건
* 해당 NaN값을 어떻게 처리해야 할 지 차후 논의 (단순히 0교체 or 주변값 or interpolate or 다른년도 동월 비교 등)

```python
template['신고건수'].isna().value_counts()
```

* '0:월, 1:화, ... ,6:일' 로 요일 컬럼 생성 

```python
template['요일']=template['date'].dt.dayofweek
template.head()
```

* 공휴일 여부 컬럼 생성

```python
cond1=template['요일']==5
cond2=template['요일']==6
template.loc[(cond1|cond2), '공휴일']=1
template.loc[~(cond1|cond2), '공휴일']=0
template.head()
```

* 주 단위 Week 컬럼 생성

```python
template['Week']=template['date'].dt.isocalendar().week
template.tail()
```

* 요일 컬럼을 '표현하고 싶은 단어의 인덱스에 1의 값을 부여하고, 다른 인덱스에는 0을 부여'하는 원핫인코딩 방식으로 재생성

```python
template=pd.get_dummies(template, columns=['요일'])
template=template.rename(columns={'요일_0':'월요일','요일_1':'화요일','요일_2':'수요일',
                      '요일_3':'목요일','요일_4':'금요일','요일_5':'토요일','요일_6':'일요일'})
template.head()
```

* 기타 특이 기간 데이터를 불러와 정기적/비정기적으로 분류 후 기존과 merge

```python
special=pd.read_csv('특이기간.csv', encoding='cp949')
special.head()
```

```python
total['정기적_이벤트']=0
total['비정기적_이벤트']=0
```

```python
for i in range(6):
    start=total['date']>=special['start_date'][i]
    end=total['date']<=special['end_date'][i]
    total.loc[(start&end),'정기적_이벤트']=1
total['정기적_이벤트'].value_counts()
```

```python
for i in range(6,13):
    start=total['date']>=special['start_date'][i]
    end=total['date']<=special['end_date'][i]
    total.loc[(start&end),'비정기적_이벤트']=1
total['비정기적_이벤트'].value_counts()
```
