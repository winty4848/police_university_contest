```python
KP20=pd.read_csv("KP2020.csv", encoding='cp949')
KP21=pd.read_csv("KP2021.csv", encoding='cp949')
NPA20=pd.read_csv("NPA2020.csv", encoding='cp949')
```

```python
KP20.columns=['접수부서코드','접수완료일시','경찰청구분','사건처리상태','사건','신고자성별','발생주소','경도','위도','동일사건여부']
KP21.columns=['접수부서코드','접수완료일시','경찰청구분','사건처리상태','사건','신고자성별','발생주소','경도','위도','동일사건여부']
NPA20.columns=['접수완료일자','접수완료시간','경찰청구분','사건처리상태','사건','신고자성별','발생주소','경도','위도','동일사건여부']
```

# KP20, KP21 접수완료일시 처리
- 데이터 후반의 '.000000000'과 같은 0 무더기 삭제
- from object to time series

```python
from datetime import datetime
KP20['접수완료일시']=KP20['접수완료일시'].str.replace('.000000000','').apply(lambda x: datetime.strptime(x,'%y/%m/%d %H:%M:%S'))
```

```python
KP20.head(3)
```

```python
KP21['접수완료일시']=KP21['접수완료일시'].str.replace('.000000000','').apply(lambda x: datetime.strptime(x,'%y/%m/%d %H:%M:%S'))
```

```python
KP21.head(3)
```

# NPA20 날짜관련 처리 (일자+시간 작업)

```python
NPA20.head(3)
```

```python
NPA20['접수완료일자']=NPA20['접수완료일자'].astype('str')
NPA20['접수완료시간']=NPA20['접수완료시간'].astype('str')
```

* 글자수 작업을 위한 str 타입으로의 변환

```python
NPA20.info()
```

```python
def make_6(x):
    return '0'*(6-len(x))+x
```

```python
NPA20['접수완료시간']=NPA20['접수완료시간'].apply(lambda x: make_6(x))
NPA20.head(3)
```

```python
NPA20['접수완료일시']=NPA20['접수완료일자']+NPA20['접수완료시간']
NPA20['접수부서코드']=np.nan
NPA20.head(3)
```

* NPA20과 KP계열 데이터의 컬럼명 맞추기 위한 작업
 - 접수부서코드는 차후 데이터 프레임 합을 위해 더미컬럼으로 붙여둠.

```python
sorting_name=['접수부서코드','접수완료일시']
for i in list(NPA20.columns[2:-2]):
    sorting_name.append(i)
sorting_name
```

```python
list(KP20.columns)==sorting_name
```

* KP20 컬럼명 순서와 sorting_name 컬럼명 순서 확인

```python
NPA20=NPA20[sorting_name]
```

```python
NPA20.head(3)
```

```python
NPA20['접수완료일시']=NPA20['접수완료일시'].apply(lambda x: datetime.strptime(x,'%Y%m%d%H%M%S'))
NPA20.head(3)
```

* 데이터 프레임 합을 위한 형식 통합 완료

```python
NPA20.info()
```

* 테스트를 위한 2020년 10월 데이터 추출

```python
NPA20[(NPA20['접수완료일시'].dt.year==2020) & (NPA20['접수완료일시'].dt.month==10)].head()
```

```python
total_df=pd.concat([NPA20,KP20,KP21],ignore_index=True)
total_df.head()
```

```python
total_df_drop_c=total_df.drop(['접수부서코드','사건처리상태','동일사건여부'],axis=1)
total_df_drop_c.head()
```

```python
total_df_drop_c['경찰청구분'].value_counts()
```

```python
cond1=total_df_drop_c['경찰청구분']==19
cond2=total_df_drop_c['경찰청구분']==13
cond3=total_df_drop_c['경찰청구분']==31
len(total_df_drop_c[cond1|cond2|cond3]) / len(total_df_drop_c)
```

* 대전/충남/세종 이외의 나머지 경찰청 다 제거 -> 약 1% 개수 감소

```python
total_df_drop_cr=total_df_drop_c[cond1|cond2|cond3]

cond4=total_df_drop_c['사건']==215
total_df_drop_cr=total_df_drop_cr[cond4]
total_df_drop_cr=total_df_drop_cr.reset_index(drop=True)
total_df_drop_cr.head()
```

```python
total_df_drop_cr.info()
```
